# Utilizando GroupBy

```python
import numpy as np
import pandas as pd
```

```python
file="C:\\projetos\\drinks.csv"
drinks = pd.read_csv(file)
drinks.head()
```

```python
#Calculando a mdia de cerveja por todo o Dataset
drinks.beer_servings.mean()
```

```python
#Calculando a media de cerveja consumida somente pelo continente Europeu
drinks[drinks.continent=='Europe'].beer_servings.mean()
```

# Mas quem bebe mais?
## Podemos calcular a mdia de cerveja por cada continente?



```python
cerveja_continente = drinks.groupby('continent').beer_servings.mean()
```

```python
#Ordenando
cerveja_continente.sort_values(ascending=False)
```

### Qual o mximo de cerveja consumida nos continentes?

```python
#Funes de agregao como min e max tambm funcionam com groupby

drinks.groupby('continent').beer_servings.max()
```

```python
#Podemos realizar tambm mltiplas agregaes com a funo agg que recebe uma lista de parametros presentes na funo describe

drinks.groupby('continent').beer_servings.agg(['count', 'mean', 'min', 'max'])
```

```python
#Caso eu no especifique a coluna, a funo groupby trar todas
drinks.groupby('continent').mean()
```

# Vamos plotar nosso primeiro grfico?

```python
%matplotlib inline
```

```python
#Agrupando as bebidas por continente, tirando a mdia e plotando um grfico de barras
drinks.groupby('continent').mean().plot(kind='bar')
```

# Exerccios

- Leia o Dataset de ordens de sanduche
- Calcule o valor das contas agrupando por order_id
- Calcule a quantidade de itens agrupando por order_id

```python
arquivo = "C:\\projetos\\sanduiches.txt"
vendas = pd.read_table(arquivo)
vendas.head()
```

```python
vendas['item_value'] = vendas.item_price.str.replace('$', '').astype(float)
```

```python
vendas.head()
```

```python
vendas_agr = vendas.groupby('order_id').item_value.sum()
```

```python
vendas_agr.head()
```

```python
qtd_agr = vendas.groupby('order_id').quantity.sum()
```
