Quando **referenciamos linhas e colunas** com o parâmetro axis:

- **axis 0** = linhas
- **axis 1** = colunas

```python
#Calcule a média de cada coluna numérica
drinks.mean()
```

```python
#Ou podemos fazer explicitando o eixo
drinks.mean(axis=0)
```

```python
drinks.describe()
```

```python
# Calculando a média para cada linha
drinks.mean(axis=1).head()
```

```python
# 'index'é um alias para axis 0
drinks.mean(axis='index')
```

```python
# 'columns' é um alias para axis 1
drinks.mean(axis='columns').head()
```
