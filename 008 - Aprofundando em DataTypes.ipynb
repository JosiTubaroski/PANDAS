```python
arquivo = "C:\\projetos\\sanduiches.txt"
vendas = pd.read_table(arquivo)
vendas.head()
```

# O preo do sanduche  uma string e est com o caracter $

### Como podemos converter em float e trazer a mdia com uma linha de cdigo? Com astype!



```python
#Convertendo uma string em um float e trazendo a mdia

vendas.item_price.str.replace('$', '').astype(float).mean()
```

# Mais um pouco sobre booleanos

```python
#J aprendemos a utilizar o contains, certo?

vendas.item_name.str.contains('Chicken').head(10)
```

```python
#Vamos converter uma serie boolean em integer (False = 0, True = 1)

vendas.item_name.str.contains('Chicken').astype(int).head(20)
```

```python
#Podemos fazer o procedimento com uma varivel?

vendas_int = vendas.item_name.str.contains('Chicken').astype(int)

vendas_int.head(20)
```

# E afinal, quantos sanduches contem Tomato?

```python
vendas.item_name.str.contains('Tomato').astype(int).sum()
```

```python
#Como  trabalhada a programao acima?
# Passo 01 - vendas.item_name.str.contains('Tomato').astype(int) TRAZ UMA SRIE BOOL CONVERTIDA EM INT
# Passo 02 - Somente aplicamos o SUM() na srie inteira, que somar as ocorrencias verdadeiras, ou seja, 1

passo01 = vendas.item_name.str.contains('Tomato').astype(int)
passo02 = passo01.sum()
passo02
```
