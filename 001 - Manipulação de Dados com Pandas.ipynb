Series com Dicionrios

```python
#Criando uma srie atravs de um dicionrio -  As chaves passa a ndices

world_cup = {'Brasil':5,'Alemanha':4,'Itlia':4,'Argentina':2,'Uruguai':2,'Inglaterra':1,'Frana':1}
```

```python
world_cup_sr = pd.Series(world_cup)
world_cup_sr
```

```python
#Criando uma srie definindo o ndice

world_cup = {'Brasil':5,'Alemanha':4,'Itlia':4,'Argentina':2,'Uruguai':2,'Inglaterra':1,'Frana':1}
```

```python
world_cup_sr = pd.Series(
    world_cup, 
    index = ['Brasil','Alemanha','Itlia','Argentina','Uruguai','Inglaterra','Espanha'], 
    name = 'wcSeries')
```

```python
#Observe o valor nulo 

world_cup_sr
```

```python
world_cup_sr.index
```

```python
#Posso trazer o valor de uma srie atravs do ndice

world_cup_sr['Brasil']
```

```python
#Implicitamente, porm, o Python ainda atribui um ndice numrico

world_cup_sr[0]
```

```python
#Podemos tambm atribuir um valor atravs do ndice numrico ou valorado

world_cup_sr['Espanha'] = 1
world_cup_sr
```

```python
#Ao trabalharmos com ndices numricos, podemos fazer slices assim como nas listas e vetores

world_cup_sr[1:3]
```

```python
#Podemos realizar comparaes lgicas

world_cup_sr[world_cup_sr > 3]
```
