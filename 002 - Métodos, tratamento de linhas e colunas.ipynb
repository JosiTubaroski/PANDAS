## Renomeando colunas no Pandas

```python
filmes.columns
```

```python
filmes.rename(columns = {'star rating':'star_rating','content rating':'content_rating'})
```

```python
filmes.columns
```

A transformao no vai ocorrer, pois estamos modificando o dataframe original e o Pandas no permite essa operao, a no ser que ela seja explcita, ou seja, devemos forar a alterao

    1 - Atribuir o resultado a uma nova variavel, ou seja, um novo Dataframe
    2 - Utilizar o argumento inplace=True forando a alterao


```python
filmes.rename(columns = {'star rating':'star_rating','content rating':'content_rating'},inplace=True)
```

```python
filmes.columns
```

## Real World

 muito comum termos arquivos com vrias colunas e a operao de renomear uma a uma fica invivel. Podemos ento, utilizar funes do Python para trabalharmos com as colunas.

```python
file="C:\\projetos\\movies.csv"
filmes = pd.read_csv(file)
filmes.head()

filmes.columns = filmes.columns.str.replace(' ', '_')
filmes.columns
```

## Removendo Colunas do Dataframe

```python
#Por boa prtica, ao realizar alguma alterao no Dataframe, devemos atribuir a outra varivel,
#assim  possvel rastrear as mudanas e tambm voltar ao ltimo passo.
#Aqui tambm podemos utilizar o inplace=True

#Axis = 1 - Colunas / Axis = 0 - Linhas

#Filmes01 - foi criado sem a coluna content_rating

filmes01 = filmes.drop('content_rating', axis=1)
filmes01.head()
```

```python
filmes.head()
```

```python
#Removendo mais de uma coluna

filmes02 = filmes01.drop(['duration', 'actors_list'], axis=1)
filmes02.head()
```

```python
filmes01.head()
```
